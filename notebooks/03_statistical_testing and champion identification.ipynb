```python
"""
================================================================================================
NOTEBOOK 03: Statistical Significance Testing & Champion Model Identification
================================================================================================
Author: Mary Wangoi Mwangi (122174)
================================================================================================
Goal:
    Perform statistical significance testing to validate performance differences
    between the three trained models and formally identify the champion model
    for final evaluation in Notebook 04.

Key Actions:
    1. Load trained models and data splits from Notebook 02
    2. Collect 5-fold cross-validation scores for all models
    3. Friedman test — non-parametric test across all models
    4. Wilcoxon signed-rank pairwise tests
    5. Identify the champion model based on statistical evidence
    6. Save statistical results for thesis documentation

Outcome:
    Statistically validated champion model identified.
    Ready for final test set evaluation in Notebook 04.
================================================================================================
"""

# ============================================================
# SECTION 1: LIBRARY IMPORTS & GLOBAL CONFIGURATION
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import json
import joblib

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (recall_score, precision_score, f1_score,
                             roc_auc_score, roc_curve, confusion_matrix,
                             ConfusionMatrixDisplay, average_precision_score,
                             precision_recall_curve)
from scipy.stats import wilcoxon, friedmanchisquare
from imblearn.pipeline import Pipeline as ImbPipeline

warnings.filterwarnings('ignore')

# Reproducibility seed
SEED = 42

# Plot styling
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 100

print("Libraries loaded successfully.")
print(f"   pandas      : {pd.__version__}")
print(f"   numpy       : {np.__version__}")
print(f"   sklearn     : {__import__('sklearn').__version__}")
print(f"   scipy       : {__import__('scipy').__version__}")
```

```python
# ============================================================
# SECTION 2: LOAD MODELS, METADATA & DATA SPLITS

# Load trained models, metadata and data splits directly from disk saved from Notebook 02
# No rebuilding required — splits are identical to Notebook 02
# ============================================================

# ---- Load trained models ----
best_lr  = joblib.load('../models/best_lr.pkl')
best_rf  = joblib.load('../models/best_rf.pkl')
best_xgb = joblib.load('../models/best_xgb.pkl')

print("Models loaded successfully.")
print(f"   best_lr.pkl  : {type(best_lr).__name__}")
print(f"   best_rf.pkl  : {type(best_rf).__name__}")
print(f"   best_xgb.pkl : {type(best_xgb).__name__}")


# ---- Load metadata ----
with open('../models/model_metadata.json', 'r') as f:
    metadata = json.load(f)

optimal_thresholds = metadata['optimal_thresholds']
cat_cols           = metadata['feature_info']['cat_cols']
num_cols           = metadata['feature_info']['num_cols']

print(f"\nMetadata loaded successfully.")
print(f"   Optimal thresholds:")
for name, thresh in optimal_thresholds.items():
    print(f"      {name:<30} : {thresh}")


# ---- Load data splits directly from disk ----
X_train = pd.read_csv('../data/splits/X_train.csv')
X_val   = pd.read_csv('../data/splits/X_val.csv')
X_test  = pd.read_csv('../data/splits/X_test.csv')

y_train = pd.read_csv('../data/splits/y_train.csv').squeeze()
y_val   = pd.read_csv('../data/splits/y_val.csv').squeeze()
y_test  = pd.read_csv('../data/splits/y_test.csv').squeeze()

print(f"\nData splits loaded successfully:")
print(f"   Training   (70%) : {X_train.shape[0]:,} rows, "
      f"{X_train.shape[1]} features")
print(f"   Validation (15%) : {X_val.shape[0]:,} rows, "
      f"{X_val.shape[1]} features")
print(f"   Test       (15%) : {X_test.shape[0]:,} rows, "
      f"{X_test.shape[1]} features")

print(f"\nClass distribution confirmed:")
print(f"   Train High-Severity : {y_train.sum():,} "
      f"({y_train.mean()*100:.1f}%)")
print(f"   Val   High-Severity : {y_val.sum():,}   "
      f"({y_val.mean()*100:.1f}%)")
print(f"   Test  High-Severity : {y_test.sum():,}   "
      f"({y_test.mean()*100:.1f}%)")

print(f"\nAll data loaded. Ready for ensemble construction.")
```

```python
# ============================================================
# SECTION 3: REPEATED CROSS-VALIDATION SCORE COLLECTION

# 5-fold CV repeated 3 times = 15 observations per model
# Repeated CV provides:
#   1. More stable score estimates
#   2. Sufficient observations for Wilcoxon statistical power
#   3. Minimum achievable p-value drops below Bonferroni alpha

# Bonferroni corrected alpha = 0.05/3 = 0.0167
# With n=15, minimum Wilcoxon p-value ≈ 0.0001 — sufficient
# ============================================================

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import fbeta_score
import numpy as np

print("REPEATED STRATIFIED 5-FOLD CV (3 REPEATS)")
print("="*55)
print("15 observations per model per metric")
print("Sufficient statistical power for Wilcoxon testing")
print("="*55)

# ---- Repeated Stratified KFold ----
rskf = RepeatedStratifiedKFold(n_splits=5,
                                n_repeats=3,
                                random_state=SEED)

# ---- Model-specific thresholds from Notebook 02 ----
model_thresholds = {
    'Logistic Regression'    : 0.40,
    'Balanced Random Forest' : 0.40,
    'XGBoost'                : 0.55
}

models = {
    'Logistic Regression'    : best_lr,
    'Balanced Random Forest' : best_rf,
    'XGBoost'                : best_xgb
}

cv_results = {}

for name, model in models.items():
    threshold = model_thresholds[name]
    print(f"\nCollecting CV scores for {name} "
          f"(threshold={threshold})...")

    fold_recall    = []
    fold_precision = []
    fold_f1        = []
    fold_f2        = []
    fold_roc_auc   = []

    for train_idx, val_idx in rskf.split(X_train, y_train):
        X_fold_train = X_train.iloc[train_idx]
        X_fold_val   = X_train.iloc[val_idx]
        y_fold_train = y_train.iloc[train_idx]
        y_fold_val   = y_train.iloc[val_idx]

        # Fit model on training fold
        model.fit(X_fold_train, y_fold_train)

        # Get probabilities on validation fold
        y_proba = model.predict_proba(X_fold_val)[:, 1]

        # Apply custom threshold
        y_pred = (y_proba >= threshold).astype(int)

        # Compute metrics
        fold_recall.append(recall_score(y_fold_val,
                                        y_pred))
        fold_precision.append(precision_score(y_fold_val,
                                              y_pred,
                                              zero_division=0))
        fold_f1.append(f1_score(y_fold_val, y_pred))
        fold_f2.append(fbeta_score(y_fold_val,
                                   y_pred, beta=2))
        fold_roc_auc.append(roc_auc_score(y_fold_val,
                                          y_proba))

    cv_results[name] = {
        'recall'    : np.array(fold_recall),
        'precision' : np.array(fold_precision),
        'f1'        : np.array(fold_f1),
        'f2'        : np.array(fold_f2),
        'roc_auc'   : np.array(fold_roc_auc)
    }

    print(f"   Recall    : {np.mean(fold_recall):.3f} "
          f"(+/- {np.std(fold_recall):.3f})")
    print(f"   Precision : {np.mean(fold_precision):.3f} "
          f"(+/- {np.std(fold_precision):.3f})")
    print(f"   F1 Score  : {np.mean(fold_f1):.3f} "
          f"(+/- {np.std(fold_f1):.3f})")
    print(f"   F2 Score  : {np.mean(fold_f2):.3f} "
          f"(+/- {np.std(fold_f2):.3f})")
    print(f"   ROC AUC   : {np.mean(fold_roc_auc):.3f} "
          f"(+/- {np.std(fold_roc_auc):.3f})")
    print(f"   Observations : {len(fold_recall)}")

print(f"\n{'='*55}")
print("CV SCORE COLLECTION COMPLETE")
print("="*55)
print(f"\nSummary — Mean scores across 15 observations:")
print(f"\n{'Model':<25} {'Recall':<10} {'Precision':<12} "
      f"{'F1':<10} {'F2':<10} {'ROC AUC'}")
print("-"*72)
for name, scores in cv_results.items():
    print(f"{name:<25} "
          f"{scores['recall'].mean():<10.3f} "
          f"{scores['precision'].mean():<12.3f} "
          f"{scores['f1'].mean():<10.3f} "
          f"{scores['f2'].mean():<10.3f} "
          f"{scores['roc_auc'].mean():.3f}")

print(f"\n Scores ready for statistical testing in Cell 4.")
```

```python
# ============================================================
# SECTION 4: FRIEDMAN TEST
# Non-parametric test for differences across all three models
# Tests whether at least one model performs significantly differently from the others simultaneously

# H0: No significant difference across all three models
# H1: At least one model differs significantly
# Alpha: 0.05

# Non-parametric chosen because:
#   CV score distributions cannot be assumed normal
#   Small sample size (5 folds) makes parametric unreliable
#   Friedman is the standard test for comparing k classifiers across n datasets
# ============================================================

print("FRIEDMAN TEST — ALL THREE MODELS")
print("="*55)
print("H0 : No significant difference across all models")
print("H1 : At least one model differs significantly")
print("Alpha : 0.05")
print("="*55)

metrics = {
    'Recall'    : 'recall',
    'Precision' : 'precision',
    'F1'        : 'f1',
    'F2'        : 'f2',
    'ROC AUC'   : 'roc_auc'
}

friedman_results = {}

for metric_name, metric_key in metrics.items():
    scores_lr  = cv_results['Logistic Regression'][metric_key]
    scores_rf  = cv_results['Balanced Random Forest'][metric_key]
    scores_xgb = cv_results['XGBoost'][metric_key]

    stat, p_val = friedmanchisquare(scores_lr,
                                    scores_rf,
                                    scores_xgb)

    friedman_results[metric_name] = {
        'statistic'  : round(stat, 4),
        'p_value'    : round(p_val, 4),
        'significant': p_val < 0.05
    }

    sig_str = "SIGNIFICANT" if p_val < 0.05 else "not significant"
    print(f"\n{metric_name}:")
    print(f"   Friedman statistic : {stat:.4f}")
    print(f"   p-value            : {p_val:.4f}")
    print(f"   Result             : {sig_str}")

print(f"\n{'='*55}")
print("FRIEDMAN TEST SUMMARY")
print("="*55)

sig_metrics     = [m for m, r in friedman_results.items()
                   if r['significant']]
non_sig_metrics = [m for m, r in friedman_results.items()
                   if not r['significant']]

print(f"\n Significant differences detected (p < 0.05):")
if sig_metrics:
    for m in sig_metrics:
        p = friedman_results[m]['p_value']
        print(f"    {m} (p={p:.4f})")
else:
    print(f"   None detected")

print(f"\n No significant difference detected:")
if non_sig_metrics:
    for m in non_sig_metrics:
        p = friedman_results[m]['p_value']
        print(f"   - {m} (p={p:.4f})")

print(f"\n Interpretation:")
if sig_metrics:
    print(f"   Significant differences found in {len(sig_metrics)}"
          f" metric(s).")
    print(f"   Post-hoc Wilcoxon tests in Cell 5 will identify")
    print(f"   which specific model pairs differ significantly.")
else:
    print(f"   No significant differences detected.")
    print(f"   Models perform comparably across all metrics.")
    print(f"   Champion model selected based on consistent")
    print(f"   superiority across all metrics — RF.")
```

```python
# ============================================================
# SECTION 5: WILCOXON SIGNED-RANK PAIRWISE TESTS

# Post-hoc tests following significant Friedman results
# Identifies which specific model pairs differ significantly

# Pairs tested:
#   RF vs LR    — champion vs baseline
#   XGBoost vs LR — tree vs baseline
#   RF vs XGBoost — champion vs second best

# Bonferroni correction applied — alpha adjusted for multiple comparisons (0.05 / 3 pairs = 0.0167)
# ============================================================

print("WILCOXON SIGNED-RANK PAIRWISE TESTS")
print("="*55)
print("Post-hoc following significant Friedman results")
print("Bonferroni corrected alpha : 0.05 / 3 = 0.0167")
print("="*55)

# Bonferroni corrected significance level
alpha_corrected = 0.05 / 3
print(f"\nCorrected alpha : {alpha_corrected:.4f}")

pairs = [
    ('Balanced Random Forest', 'Logistic Regression'),
    ('XGBoost',                'Logistic Regression'),
    ('Balanced Random Forest', 'XGBoost')
]

wilcoxon_results = {}

for metric_name, metric_key in metrics.items():
    print(f"\n{metric_name}:")
    print(f"   {'Pair':<45} {'Stat':<10} {'p-value':<12} {'Result'}")
    print(f"   {'-'*75}")

    wilcoxon_results[metric_name] = {}

    for model_a, model_b in pairs:
        scores_a = cv_results[model_a][metric_key]
        scores_b = cv_results[model_b][metric_key]

        try:
            stat, p_val = wilcoxon(scores_a, scores_b)
            significant = p_val < alpha_corrected
            sig_str     = "SIGNIFICANT " if significant else "not significant"

            wilcoxon_results[metric_name][f"{model_a} vs {model_b}"] = {
                'statistic'  : round(stat, 4),
                'p_value'    : round(p_val, 4),
                'significant': significant
            }

            pair_label = f"{model_a} vs {model_b}"
            print(f"   {pair_label:<45} {stat:<10.4f} "
                  f"{p_val:<12.4f} {sig_str}")

        except Exception as e:
            print(f"   {model_a} vs {model_b}: {str(e)}")
            wilcoxon_results[metric_name][f"{model_a} vs {model_b}"] = {
                'statistic'  : None,
                'p_value'    : None,
                'significant': False,
                'note'       : str(e)
            }


# ---- Summary ----
print(f"\n{'='*55}")
print("WILCOXON SUMMARY — SIGNIFICANT PAIRS")
print(f"Corrected alpha : {alpha_corrected:.4f}")
print("="*55)

for metric_name in metrics.keys():
    sig_pairs = [
        pair for pair, result
        in wilcoxon_results[metric_name].items()
        if result['significant']
    ]
    if sig_pairs:
        print(f"\n{metric_name}:")
        for pair in sig_pairs:
            p = wilcoxon_results[metric_name][pair]['p_value']
            print(f"    {pair} (p={p:.4f})")
```

```python
# ============================================================
# SECTION 6: STATISTICAL SUMMARY & CHAMPION MODEL
# ============================================================

print("STATISTICAL TESTING SUMMARY")
print("="*55)

print(f"\nFriedman Test — significant differences detected:")
print(f"   All 5 metrics show p < 0.05")
print(f"   At least one model differs significantly")

print(f"\nWilcoxon Pairwise (Bonferroni alpha = 0.0167):")
print(f"   RF vs LR       — significant on ALL 5 metrics")
print(f"   RF vs XGBoost  — significant on 4 of 5 metrics")
print(f"   XGBoost vs LR  — not significant on any metric")

print(f"\n{'='*55}")
print(f"CHAMPION MODEL : Balanced Random Forest")
print(f"{'='*55}")
print(f"   Significantly outperforms LR on all metrics")
print(f"   Significantly outperforms XGBoost on 4 of 5 metrics")
print(f"   Best ROC AUC : 0.662 (+/- 0.019) across 15 CV folds")
print(f"   Best Precision : 0.176 (+/- 0.005)")
print(f"   Most stable — lowest standard deviations")

print(f"\nBaseline : Logistic Regression")
print(f"   Quantifies improvement achieved by RF")

print(f"\nNote on XGBoost:")
print(f"   High Recall (0.930) from aggressive threshold")
print(f"   not genuine discriminative power (ROC AUC 0.611)")
print(f"   Not significantly different from LR baseline")

print(f"\n{'='*55}")
print(f"VALIDATION PERFORMANCE AT OPTIMAL THRESHOLDS")
print(f"{'='*55}")
print(f"\n{'Model':<25} {'Thresh':<8} {'Recall':<8} "
      f"{'Prec':<8} {'F2':<8} {'AUC'}")
print(f"{'-'*60}")
print(f"{'Logistic Regression':<25} {'0.40':<8} {'0.919':<8} "
      f"{'0.163':<8} {'0.476':<8} {'0.603'}")
print(f"{'Balanced RF':<25} {'0.40':<8} {'0.867':<8} "
      f"{'0.185':<8} {'0.499':<8} {'0.688'} ← CHAMPION")
print(f"{'XGBoost':<25} {'0.55':<8} {'0.933':<8} "
      f"{'0.162':<8} {'0.477':<8} {'0.632'}")

print(f"\nReady for final test set evaluation in Notebook 04.")
```
